## Indice de Correlação

```python
# Mapa de calor de correlação
plt.figure(figsize=(10, 8))
corr = df[indicadores].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação entre Indicadores')
plt.show()
```

## A análise da **Matriz de Correlação entre Indicadores** 

### 1. Protagonistas do INDE (Correlações Fortes)

O **INDE** não é influenciado por todos os indicadores da mesma forma. Três variáveis se destacam como as que mais "empurram" o índice geral para cima ou para baixo:

* **IDA (0.78):** É o indicador com a maior correlação. Isso confirma que o desempenho acadêmico (notas) é o fator determinante para a classificação final do aluno.
* **IPV (0.72):** O "Ponto de Virada" tem uma influência altíssima, mostrando que a percepção pedagógica de evolução está muito alinhada ao sucesso geral no programa.
* **IEG (0.58):** O engajamento também possui uma relação sólida e positiva com o INDE.

### 2. Relações Inter-Indicadores (Sinergias)

Observando a matriz fora da linha do INDE, notamos conexões interessantes entre as dimensões:

* **IDA vs. IPV (0.56):** Existe uma forte sinergia entre boas notas e a identificação do ponto de virada pelos professores.
* **IEG vs. IPV (0.44):** Alunos mais engajados (lição de casa/voluntariado) tendem a ser vistos com maior potencial de evolução pedagógica.
* **IPP vs. IPV (0.52):** Os indicadores que vêm do "Conselho" (pedagogos e professores) andam de mãos dadas, mostrando consistência na avaliação da Dimensão Psicopedagógica.

### 3. As "Ilhas" de Independência (Correlações Fracas)

Alguns indicadores parecem operar de forma quase independente dos outros:

* **IPS (Indicador Psicossocial):** Apresenta as correlações mais baixas da matriz (ex: **0.03** com IDA e **-0.05** com IPV). Isso indica que os aspectos psicológicos avaliados não dependem necessariamente de o aluno tirar notas boas ou ser engajado.
* **IAA (Autoavaliação):** Também apresenta baixa correlação com os indicadores de desempenho acadêmico (**0.12** com IDA). O que o aluno pensa de si mesmo não está fortemente vinculado à sua nota real em prova.

---

### Diagnóstico Estratégico

* **Foco no IDA:** Como o IDA tem a correlação mais alta com o INDE (**0.78**), qualquer projeto de reforço escolar que aumente as notas terá um impacto imediato e proporcional na melhoria das "Pedras-conceito" da instituição.
* **A Independência do IPS:** O fato de o IPS ser "independente" é positivo; ele funciona como um radar para problemas que as notas não mostram (um aluno excelente pode ter um IPS baixo e precisar de suporte emocional).

## Cruzando dados de DEFA com os outliers

```python
# Criando a coluna de status baseada no valor de DEFA
df['Status_DEFA'] = df['DEFA'].apply(
    lambda x: 'Atrasado' if x < 0 else ('No Prazo' if x == 0 else 'Adiantado')
)

# Verificando a contagem de cada grupo
print(df['Status_DEFA'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))
sns.boxplot(data=df, x='FASE', y='INDE', hue='Status_DEFA', 
            hue_order=['Atrasado', 'No Prazo', 'Adiantado'], 
            palette='coolwarm')

plt.title('Impacto da Defasagem (DEFA) no INDE por Fase')
plt.ylabel('Índice de Desenvolvimento Educacional (INDE)')
plt.xlabel('Fase de Ensino')
plt.legend(title='Status de Defasagem', loc='upper right')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Filtrando apenas alunos com INDE considerado baixo (ex: abaixo de 5) e vendo o DEFA médio
baixo_desempenho = df[df['INDE'] < 5]
print("Perfil dos alunos com baixo INDE:")
print(baixo_desempenho['Status_DEFA'].value_counts(normalize=True) * 100)
```

# Proposta de trabalho baseada no EDA

Com base na análise técnica que realizamos e no perfil dos dados da sua base (foco em indicadores contínuos, forte correlação com o **IDA** e o impacto crítico da defasagem), o modelo de predição mais indicado para a escola é um de **Classificação de Risco de Desempenho** (utilizando algoritmos como **Random Forest** ou **XGBoost**).

Abaixo está a proposta detalhada do modelo:

### 1. Objetivo do Modelo: Predição de "Pedra-Conceito" e Alerta de Risco

Em vez de apenas prever um número, o modelo deve classificar o aluno em tempo real para identificar quem está em risco de cair para a categoria **Quartzo** ou **Ágata**.

* **Variável Alvo (Target):** A categoria da "Pedra-Conceito" (ou uma variável binária: "Em Risco" vs. "Desempenho Sólido").
* **Variáveis Preditoras (Features):**
* **Histórico de Notas (IDA):** Por ter a maior correlação (**0.78**), é o preditor mais forte.
* **Status de Defasagem (Status_DEFA):** Como **80%** dos alunos de baixo INDE são atrasados, essa variável é um sinalizador precoce de risco.
* **Engajamento (IEG):** Para identificar se o aluno está "esforçado" mas não está conseguindo notas.
* **Ponto de Virada (IPV):** Utilizar a percepção do professor como um dado qualitativo pesado no modelo.



### 2. Algoritmo Recomendado: Random Forest Classifier

Este modelo é ideal para o seu cenário por ser robusto a *outliers* (como os que vimos nos boxplots) e por lidar bem com variáveis de diferentes escalas.

* **Vantagem Pedagógica:** Ele fornece a "Importância das Variáveis", permitindo que a escola saiba exatamente o que está pesando mais para um aluno estar "atrasado".

### 3. Aplicação Prática: O "Semáforo Pedagógico"

Com o modelo treinado, a escola pode implementar um sistema de intervenção preventiva:

| Alerta | Ação Sugerida | Baseado em |
| --- | --- | --- |
| **Crítico (Quartzo)** | Plano de aceleração imediata e apoio psicológico. | Baixo IDA + Status Atrasado. |
| **Atenção (Ágata)** | Reforço acadêmico focado em aumentar o IDA. | Engajamento (IEG) alto, mas notas baixas. |
| **Potencial (Ametista)** | Mentoria para busca de excelência (Topázio). | Estabilidade no IPV e IPP. |
